```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

# Importo los datos

```python
Si = pd.read_csv( 'Si.txt',decimal=',', header=0, names= ["temperatura_Si","energia_Si"], delim_whitespace=True )

Ge = pd.read_csv( 'Ge.txt',decimal=',', header=0, names= ["temperatura_Ge","energia_Ge"], delim_whitespace=True )

Diodo1 = pd.read_csv( 'Diodo1.txt',decimal=',', header=0, names= ["temperatura_Diodo1","energia_Diodo1"], delim_whitespace=True )

Diodo2 = pd.read_csv( 'Diodo2.txt',decimal=',', header=0, names= ["temperatura_Diodo2","energia_Diodo2"], delim_whitespace=True )
```

# Silicio

```python
x=Si.iloc[:,0]
y=Si.iloc[:,1]

def ajuste(x, a, b, c):
    return a+b*x+c*x*np.log(x)

popt,pcov= curve_fit(ajuste,x,y,p0=[3,-0.0015,0])
plt.plot(x,y,"b.")
plt.plot(x,ajuste(x,*popt),"r-", label="fit",alpha=0.7)
plt.legend()
plt.xlabel("Temperatura[K]")
plt.ylabel("Energía[eV]")
plt.title("Silicio")
plt.show()
print (popt[0],pcov[0])
```
