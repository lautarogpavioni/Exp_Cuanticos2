```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### importamos los datos:

```python
fondo = pd.read_csv( 'Fondo2.txt',decimal=',', header=None, names= ["canal","cuentas"], delim_whitespace=True )

hg = pd.read_csv( 'hg2.txt',decimal=',', header=None, names= ["canal","cuentas"], delim_whitespace=True )

sustrato = pd.read_csv( 'itosustrato.txt',decimal=',', header=None, names= ["canal","cuentas"], delim_whitespace=True )

ZnMnO = pd.read_csv( 'ZnMnO.txt',decimal=',', header=None, names= ["canal","cuentas"], delim_whitespace=True )

ZnMnO2 = pd.read_csv( 'ZnMnOrecocido.txt',decimal=',', header=None, names= ["canal","cuentas"], delim_whitespace=True )
```

### ploteamos para ver la forma de los datos:

```python
fondo.plot('canal','cuentas', title= "fondo")
hg.plot('canal','cuentas', title= "Hg")
sustrato.plot('canal','cuentas', title= "sustrato")
ZnMnO.plot('canal','cuentas', title= "ZnMnO")
ZnMnO2.plot('canal','cuentas', title= "ZnMnO2")
```
